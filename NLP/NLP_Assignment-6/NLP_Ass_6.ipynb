**STEPS :-**

1.Import Model – Import pipeline from the transformers library.

2.Initialize Classifier – Create a sentiment analysis pipeline using pipeline("sentiment-analysis").

3.Take User Input – Prompt the user to enter text.

4.Analyze Sentiment – Process the input text using the classifier.

5.Display Output – Print the sentiment label and confidence score.



```python
from transformers import pipeline

# Sentiment Analysis using Transformers
classifier = pipeline("sentiment-analysis")

# Input text from user
text = input("Enter text for sentiment analysis: ")

# Get sentiment result
result = classifier(text)

# Display result
print(result)
```
